```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotting(main_title, train_acc, test_acc, number_of_nodes):
    plt.figure(figsize=(12,6), dpi=100)
    plt.subplot(121)
    plt.suptitle(main_title)
    plt.title("Train accuracy")
    plt.semilogx(number_of_nodes, train_acc)
    plt.ylabel("Accuracy")
    plt.xlabel("Number of nodes")

    plt.subplot(122)
    plt.title("Test accuracy")
    plt.semilogx(number_of_nodes, test_acc)
    plt.ylabel("Accuracy")
    plt.xlabel("Number of nodes")
    plt.subplots_adjust(hspace=8, wspace=0.4)

    plt.savefig('plots/' + main_title+'.png')
```
